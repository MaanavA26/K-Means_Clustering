# Unsupervised Learning
K-Means Clustering using custom data


# Import Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Data

```python
data = pd.read_csv("C://Users//DITU//Downloads//datasets-master//datasets-master//titanic.csv")
data=data.dropna()
#data.head(900)
```

# Taking two columns for our plot

```python
x = data["Age"]
```

```python
y = data["Fare"]
```

```python
data = list(zip(x.values,y.values))
print(data)
```

# Fitting Data

```python
model = KMeans(n_clusters = 6)
model.fit(data)
```

# Plotting Data


```python
plt.scatter(x,y, c= model.labels_)
plt.show()
```
